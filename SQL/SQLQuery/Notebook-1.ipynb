```sql
select p.FullName, p.LogonName, p.EmailAddress
from Application.People p
```

```sql
select p.FullName, p.LogonName, p.EmailAddress
, CHARINDEX('@', p.EmailAddress) as POSITION

, LEFT(p.EmailAddress,CHARINDEX('@', p.EmailAddress)-1)as Mailbox
, SUBSTRING(p.EmailAddress, CHARINDEX('@', p.EmailAddress)+1,Len(p.EmailAddress))as Domain
from Application.People p;
```

```sql
SELECT SUBSTRING(Fullname, CHARINDEX('', p.FullName)+1, 1) +
 '.' + SUBSTRING(FullName, CHARINDEX(' ', p.FullName)+1, 100) as uName,
 IIF(IsSalesperson=1,'Sales Person',IIF(IsEmployee=1,'Employee','Customer')) as [Contact Type]

FROM Application.People p
;
```

```sql
SELECT SUBSTRING(Fullname, CHARINDEX('', p.FullName)+1, 1) +
 '.' + SUBSTRING(FullName, CHARINDEX(' ', p.FullName)+1, 100) as uName,

IIF(IsSalesperson=1,'Sales Person',IIF(IsEmployee=1,'Employee','Customer')) as [Contact Type]

FROM Application.People p
;
```

```sql
SELECT COUNT(IIF(p.IsSalesperson=1,1,null)) as Sales
,COUNT(IIF(p.IsEmployee=1,1,null))as Employee
,COUNT (IIF(p.IsSystemUser=1,1,null))as Other

FROM Application.People p
;
```

```sql
Select Count(o.OrderID) as OrderCount,DaysToDeliver
FROM
(SELECT OrderID, DATEDIFF(DAY,OrderDate,ExpectedDeliveryDate) as "DaysToDeliver"
,ExpectedDeliveryDAte,OrderDate
FROM Sales.Orders) o
group by DaysToDeliver
;

```
